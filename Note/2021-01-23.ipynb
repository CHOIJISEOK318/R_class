## 두개 이상의 데이터프레임 병합하기

```r
# Case 1: key값이 되는 컬럼이 동일하고, data의 갯수도 동일한 경우
height <- data.frame(id=1:3, height=c(160,170,180))
height
weight <- data.frame(id=c(2,1,3),weight=c(70,60,80))
weight

merge(height, weight)
```

```r
# Case 2: key값이 동일하지 않고, 각각의 data의 갯수도 다를경우
height <- data.frame(id=1:3, height=c(160,170,180))
height
weight <- data.frame(ad=c(2,1),weight=c(70,60))
weight

merge(height, weight)
```

```r
#key를 정의하고 : data가 적은 쪽의 기준으로 merge
merge(x = height, y = weight, by.x="id",by.y="ad")
```

```r
#data가 많은 쪽의 기준으로 merge
merge(x = height, y = weight, by.x="id",by.y="ad", all.x = T)
```

# 문자열 처리
- 수집한 데이터의 문자열을 자르고, 교체하고, 추출하는 작업

### stringr package 사용 (function)
- str_length() : 문자열 길이
- str_c() : R에서 쓰는 Combine 'c'가 아니라 connect시켜주는 'c'이다.
- str_sub() : 범위에 해당하는 문자열 생성 (똑같네?)
- str_split() : 구분자를 기준으로 문자열을 분리하여 부분 생성
    ex)"123, 456, 789"있으면 123  // 456 // 789 분리해서 생성
- str_replace() : 기존 문자열 중에 첫번째 문자열을 특정 문자열로 변환하는 거다!?( 첫번째!!!)
- str_replace_all() : 기존 문자열 중에 전체를 특정 문자열로 변환하는 거다!? ( 전체!!!)
- str_remove(), str_remove_all() : 문자열 제거                ///'all' 없으면 무조건 첫번째 것만!!!!///
- str_extract(), str_extract_all() : 특정 패턴의 문자열 추출    ///'all' 없으면 무조건 첫번째 것만!!!!///
- str_locate(), str_locate_all() : 특정 패턴의 위치 추출       ///'all' 없으면 무조건 첫번째 것만!!!!///

# 제어문(조건문)


```r
#1
x <- 10
y <- runif(1) * 100

if(x>y){
    print('x>runif()')
    print(y)
}
print('---end---')
```

```r
#2
x <- 10
y <- runif(1) * 100
if(x>y){
    print('x>runif()')
    print(y)
}else{
    print('x<=runif()')
    print(y)
}
print('---end---')
```

```r
#3
x <- 10
y <- trunc(runif(1) * 100)

if(x>y){
    print('x>runif()')
    print(y)
}else if(x<y){
    print('x<runif()')
    print(y)
}else{
    print('x=runif()')
    print(y)
}
print('---end---')
   
```

```r
# 1~100 짝수의합
sum<-0
sum2<-0

for(i in seq(2,100,2)){
    sum<-sum+i
}
cat("1~100 짝수의합은",sum,"입니다.\n")

for(i in 1:100){
    if(i %% 2 ==0){
        sum2<-sum2+i
    }
}

cat("1~100 짝수의합은",sum2,"입니다.")
```

```r

num1 <- 0
num2 <- 0
for(i in 1:length(excel$q3)){
    ifelse(excel$q3[i]%%2==0,num1<-num1+excel$q3[i],num2<-num2+excel$q3[i])
}

cat("홀수의 핪은",num2,"짝수의 합은",num1)

num1<-0
for(i in excel$q3){
    if(i %%2 == 0 ){
        num1 <- num1 + i
    }
}
num1
```

```r
# excel의 q3컬럼 데이터중 3의 배수의 합계를 구하시오
data1 <- excel$q3
sum <- 0
for(i in data1){
    if(i %% 3 == 0){
        sum = sum + i
    }
}
cat("3의 배수의 합:",sum)

```

```r

# excel의 q3에서 3의 배수도 되고 4의 배수도 되는 수의 합계

data1 <- excel$q3
sum <- 0
for(i in data1){
    if(i %% 3 == 0&&i %% 4 == 0){
        sum = sum + i
    }
}
cat("3의 배수도되고, 4의배수도 되는 수의 합:",sum)
```
