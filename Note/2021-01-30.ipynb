```r
# 자치구별 발생건수의 평균
regionAcc.mean<-aggregate(발생건수 ~ 자치구명,seoul, mean)
```

```r
# regionAcc.mean의 발생건수를 기준으로 내림차순으로 정렬하기
arrange(regionAcc.mean,desc(발생건수))
arrange(regionAcc.sd,desc(발생건수))
```

```r
regionAcc.sd<-aggregate(발생건수 ~ 자치구명,seoul,sd)
regionAcc.sd
```

```r
# 자치구별 발생건수의 변동계수
# 변동계수 = 표준편차 / 평균
regionAcc.cv <- regionAcc.sd$발생건수 / regionAcc.mean$발생건수
regionAcc.cv
```

```r
# 자치구명과 변동계수를 가진 DataFrame 만들기
regionAcc <- data.frame(자치구명 = regionAcc.mean$자치구명, 발생건변동계수 = regionAcc.cv)
regionAcc
```

```r
# head(regionAcc)
arrange(regionAcc,변동계수)
```

```r
# 서울시 교통사고 부상자수에 대한 변동계수
#평균
#표준편차
#변동계수
#dataframe
head(seoul)
```

```r
# 부상자수 평균
injuryAcc.mean<- (aggregate(부상자수 ~ 자치구명,seoul, mean))
injuryAcc.mean
```

```r
# 표준편차
injuryAcc.sd <- (aggregate(부상자수 ~ 자치구명,seoul,sd))
regionAcc.sd
```

```r
# 자치구별 부상자수 변동계수
# 변동계수 = 표준편차 / 평균
injuryAcc.cv <- injuryAcc.sd$부상자수 / injuryAcc.mean$부상자수
injuryAcc.cv
```

```r
# 자치구명과 변동계수를 가진 DataFrame 만들기
injuryAcc <- data.frame(자치구명 = injuryAcc.mean$자치구명, 부상자변동계수 = injuryAcc.cv)
injuryAcc
```

```r
# 교통사고 발생건수와 부상자수의 관계 ## 97.9% 확률로 발생
cor(seoul$발생건수, seoul$부상자수)
```
