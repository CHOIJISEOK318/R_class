```r
# 구구단 함수 만들기 -1
gugu <- function(){
    num <- as.numeric(readline("숫자를 입력하세요!"))
    for(i in 1:9){
        cat(num," * ",i," = ",num1*i,"\n")
    }
}
```

```r
# 구구단 함수 만들기 -2
gugu <- function(num){
    for(i in 1:9){
        cat(num," * ",i," = ",num*i,"\n")
    }
}
```

```r
gugu(19)
```

```r
# 구구단 함수 만들기 - 3
gugu <- function(num,num2){
    for(i in 1:num2){
        cat(num," * ",i," = ",num*i,"\n")
    }
}
```

```r
gugu(3,20)
```

```r
# 구구단 전체출력 

for(i in 2:9){
    cat("----",i," 단 ---- \n")
    for(j in 1:9){
        cat(i," * ",j," = ",j*i,"\n")
        
    }
    cat("\n")
}
```

```r
# 구구단중 짝수단과 짝수 곱해지는수만 출력하기
for(i in seq(2,9,2)){
    
    cat("----",i," 단 ---- \n")
    for(j in seq(2,9,2)){
        cat(i," * ",j," = ",j*i,"\n")
        
    }
    cat("\n")
}

for(i in 1:9){
    if(i%%2==0){
        cat("----",i," 단 ---- \n")
        for(j in 2:9){
            if(j%%2==0){
                    cat(i," * ",j," = ",j*i,"\n")
                }
        }
        cat("\n")
    }
}
```

```r
# A컬럼의 최대값, 최소값, 빈도수
max(test$A)
min(test$A)
table(test$A)
```

```r
# for 문을 이용한 각 컬럼의 빈도수 구하기

tableMaxMin <- function(test){
    cat("-------------------------\n")
    for(i in 1:length(test)){
        cat("[*",colnames(test)[i],"*]","컬럼의 빈도 분석 결과")
        print(table(test[,i]))
    
        cat("최대값 :",max(table(test[,i])),"최소값 :",min(table(test[,i])),"\n")
        cat("-------------------------\n")
    }
}
```

```r
# 각 컬럼의 빈도수의 최대값과 최소값 구하기
tableMaxMin(test)
```

# 결측치 (NA) 자료 평균 계산

```r
data <- c(10,20,5,4,40,7,NA,6,3,NA,2,NA)
```

```r
table(is.na(data))
```

```r
# 결측치를 무조건 제거하면 통계량을 얻을 수 없으며 데이터의 손실로 문제가 발생한다.
mean(data, na.rm=T)
length(data)
```

```r
# 1) NA를 0으로 대체
data <- ifelse(is.na(data),0,data)
data
mean(data)
```

```r
is.na# 2) NA를 평균으로 대체
mean(ifelse(is.na(data),mean(data,na.rm=T),data))
```

```r
# 3) NA 를 중앙값으로 대체
mean(ifelse(is.na(data),median(data,na.rm=T),data))
```

```r
# 4) NA 를 최대값으로 
mean(ifelse(is.na(data),max(data,na.rm=T),data))
```

```r
# 4) NA 를 최소값으로 대체
mean(ifelse(is.na(data),min(data,na.rm=T),data))
```

# 몬테카를로 시뮬레이션
## 동전의 앞면 뒷면 시뮬레이션

```r
x<-runif(1)
x
if(x<=1/2){
    result <- 0
}else{
    result <- 1
}
print(result)
```

```r
coin <- function(x){
    result1<-0
    result2<-0
    
    for(i in 1:x){
        if(runif(1)<=1/2){
            result1 <- result1+1
        }else{
            result2 <- result2+1
        }
    }
    cat("앞면 :",result1,"뒷면 :",result2,"\n")
}
```

```r
coin(10)
coin(100)
coin(1000)
coin(10000)
coin(100000)
```

```r

options(scipen=100)
dice <-function(n){
    a <- rep(0,6)
    cat("Throw :",n,"\n")
    num<-0
    for(i in 1:n){
        num = trunc(runif(1)*6)+1
        a[num]=a[num]+1
    }
    for(i in 1:6){
        cat(i,"번 :",a[i],"\n")
    }
    cat("-----------------\n")
}
```

```r
dice(10)
dice(100)
dice(1000)
dice(10000)
dice(100000)
```
