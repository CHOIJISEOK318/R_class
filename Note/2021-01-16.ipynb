```r
---
#자료형 확인
- is.numeric(x) : x가 수치형?
- is.logical(x) : x가 논리형?
- is.interger(x) : x가 정수형?
- is.double(x) : x가 실수형 ?
- is.data.fram(x) : x가 dataframe?
- is.factor(x) : x가 범주형 ? #R에만 있음, 범주형 : 사람이 만들어 놓은 규칙
- is.na(x) : x에 NA가 있냐 ?
```

```r
---
#자료형 변환
- as.numeric(x) : x를 수치형으로 변환
- as.interger(x) : x를 정수형으로 변환
- as.logicasl(x) : x를 논리형으로 변환
- as.double(x) : x를 실수형으로 변환
- as.data.frame : x를 dataframe으로 변환
- as.factor(x) : x를 범주형(요인형)으로 변환
- as.Date(x) : x를 날짜형으로 변환
```

```r
#자료형 확인후 숫자 벡터로 변환하기
num1 <- c(1,2,3,"4")
num1
class(num1)
num2 <- as.numeric(num1)
num2
```

```r
##요인형(범주형) 변환
# 요인(factor)는 동일한 값의 목록을 범주로 갖는 벡터 자료
# 요인형 변수는 문자와 숫자를 동시에 가지고 있어서 계산에도 사용됨
gender <- c("man","woman","woman","man","man")
gender
class(gender)
```

```r
---
# Vector 자료구조
벡터는 R에서 가장 기본이 되는 자료구조 이다.
- 1차원 선형 자료 구조 형태
- 구성 : 변수[첨자] 형태로 구성, 첨자는 1부터 시작
- 동일한 자료형의 데이터만 저장 할 수 있다.
- 벡터 생성함수 : c(), I seq(), rep(), :
- 벡터 처리 함수 : sum(), mean(),sd(), union(), ......

```
